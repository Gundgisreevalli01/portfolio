```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# 1. Generate summary statistics
summary_stats = df.describe()
print("Summary Statistics:\n", summary_stats)

# 2. Create histograms and boxplots for numeric features
numeric_features = df.select_dtypes(include=['number']).columns

for feature in numeric_features:
    # Histogram
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.show()

    # Boxplot
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()

# 3. Use pairplot and correlation matrix for feature relationships
# Pairplot for numeric features
sns.pairplot(df[numeric_features])
plt.show()

# Correlation matrix heatmap
# Only use numeric columns for correlation matrix
corr_matrix = df[numeric_features].corr()  # Changed to only use numeric features
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 4. Identify patterns, trends, or anomalies
# Visual inspections: Check histograms, boxplots, pairplots, and heatmap

# 5. Make basic feature-level inferences from visuals
# Example: 
# - Fare distribution shows skewness, potential outliers in high fares.
# - Age distribution indicates majority of passengers are young adults.
# - Correlation heatmap reveals that fare is highly correlated with class and age.
```
